#Theortical Questions


1. What is a Support Vector Machine (SVM)?

- SVM is a supervised machine learning algorithm used for classification and regression. It finds the optimal hyperplane that maximizes the margin between two classes. The decision boundary is influenced only by a few critical points called support vectors. SVM works well in high-dimensional spaces. It supports both linear and non-linear classification using kernels.

2. What is the difference between Hard Margin and Soft Margin SVM?

- Hard Margin SVM assumes perfectly separable data and does not allow any misclassification. Soft Margin SVM allows some violations using slack variables to handle noise and overlapping points. Hard Margin gives a stricter boundary but is sensitive to outliers. Soft Margin provides better generalization by balancing errors with margin. Most real applications use Soft Margin.

3. What is the mathematical intuition behind SVM?

- SVM aims to maximize the margin between the separating hyperplane and the closest points. Maximizing margin increases model robustness and reduces overfitting. It solves a convex optimization problem that minimizes ‖w‖² subject to class constraints. The solution depends only on support vectors. This makes SVM computationally efficient in its dual form.

4. What is the role of Lagrange Multipliers in SVM?

- Lagrange multipliers convert the constrained optimization into a dual problem. They help enforce margin constraints while simplifying the objective. They allow the solution to depend only on dot products between samples, enabling kernel usage. Points with non-zero multipliers become support vectors. This makes training feasible in high dimensions.

5. What are Support Vectors in SVM?

- Support vectors are points that lie closest to the decision boundary. They define the margin and directly influence the hyperplane’s position. Removing them changes the model, while removing other points may not. These vectors have non-zero Lagrange multipliers. They are critical for SVM’s efficiency and robustness.

6. What is a Support Vector Classifier (SVC)?

- SVC is the classification implementation of SVM used for binary and multi-class tasks. It finds an optimal separating hyperplane using either linear or kernel-based methods. SVC supports soft-margin tuning through the C parameter. It is effective for complex and high-dimensional data. Kernels enable non-linear boundaries.

7. What is a Support Vector Regressor (SVR)?

- SVR applies SVM principles to regression problems. It fits a function within an ε-insensitive tube, ignoring small errors. The goal is to minimize model complexity while keeping predictions within the allowed range. Like SVM, only boundary points affect the function. SVR is effective for non-linear regression with kernels.

8. What is the Kernel Trick in SVM?

- The kernel trick maps data into higher dimensions without computing the transformation explicitly. It allows SVM to create non-linear decision boundaries using simple dot-product operations. Kernels like polynomial and RBF enable flexible modeling of complex patterns. This avoids high computational cost. It is key to SVM’s power.

9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.

- Linear kernel works best for linearly separable data and is fast to compute. Polynomial kernel captures interactions by using degree-based feature combinations. RBF kernel handles highly non-linear patterns by creating localized decision boundaries. Linear is simple, polynomial is flexible, and RBF is highly powerful. RBF is most commonly used in practice.

10. What is the effect of the C parameter in SVM?

- C controls the trade-off between margin size and classification error. High C forces the model to classify points correctly, risking overfitting. Low C allows a wider margin with more misclassifications, improving generalization. It regulates how strictly SVM penalizes errors. Proper tuning is essential for performance.

11. What is the role of the Gamma parameter in RBF Kernel SVM?

- Gamma controls how far the influence of a training point reaches. High gamma makes the model sensitive and creates complex boundaries, risking overfitting. Low gamma produces smoother boundaries but may underfit. Gamma and C together determine the decision surface. Proper tuning is crucial in RBF SVM.

12. What is the Naïve Bayes classifier, and why is it called “Naïve”?

- Naïve Bayes is a probabilistic classifier based on Bayes’ Theorem. It assumes that all features are conditionally independent given the class. This assumption is rarely true, making it “naïve.” Despite this, the classifier works extremely well in practice. It is especially useful for high-dimensional data.

13. What is Bayes’ Theorem?

- Bayes’ Theorem calculates the probability of an event using prior knowledge and observed evidence. It is expressed as: P(A|B) = [P(B|A)P(A)] / P(B). It forms the foundation of Naïve Bayes classification. It allows updating beliefs when new data arrives. It is widely used in probability and statistics.

14. Differences between Gaussian NB, Multinomial NB, and Bernoulli NB.

- Gaussian NB is used for continuous features assuming normal distribution. Multinomial NB works on count-based data like term frequencies. Bernoulli NB handles binary occurrence features such as word presence. Each variant suits different feature types. Proper choice depends on dataset structure.

15. When should you use Gaussian Naïve Bayes over other variants?

- Use Gaussian NB when features are continuous and follow a roughly Gaussian distribution. It is effective in sensor data, medical measurements, and real-valued attributes. Count-based models like Multinomial NB cannot handle such data. Gaussian NB is simple and computationally efficient. It often gives strong baseline results.

16. What are the key assumptions made by Naïve Bayes?

- It assumes that all features are independent given the class label. It treats each feature as contributing equally and independently to the probability. It also assumes reliable prior and likelihood estimates. These assumptions simplify computations significantly. However, they may not hold in correlated datasets.

17. What are the advantages and disadvantages of Naïve Bayes?

- Advantages include fast training, low computational cost, and strong performance on large and sparse data. It works well even with limited training samples. Disadvantages include sensitivity to feature dependency violations and zero-frequency issues. It may produce biased results when independence assumptions fail. Laplace smoothing helps mitigate some issues.

18. Why is Naïve Bayes a good choice for text classification?

- Text features are high-dimensional and sparse, which suits Naïve Bayes well. Word occurrences often behave independently enough for the model to work effectively. It trains extremely fast, making it ideal for large corpora. It performs well in spam detection and sentiment analysis. It remains a strong baseline for NLP tasks.

19. Compare SVM and Naïve Bayes for classification tasks.

- SVM builds complex decision boundaries and often yields higher accuracy on structured and numeric datasets. Naïve Bayes is extremely fast and robust for high-dimensional data like text. SVM requires tuning parameters like C and gamma, while NB is simpler. SVM handles non-linear data better with kernels. NB scales better to large feature spaces.

20. How does Laplace Smoothing help in Naïve Bayes?

- Laplace smoothing adds a small constant (usually 1) to all feature counts. It prevents zero probabilities for unseen features during training. This ensures that one missing word does not make the entire probability zero. It improves stability in text classification models. It is essential for Multinomial and Bernoulli NB.

1. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

pipe = SVC(kernel="rbf", C=1.0, gamma="scale", random_state=42)
pipe.fit(X_tr, y_tr)
pred = pipe.predict(X_te)
print("Accuracy:", accuracy_score(y_te, pred))
```

4. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

X, y = make_moons(n_samples=400, noise=0.25, random_state=42)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, C=1.0, gamma="scale"))
clf.fit(X_tr, y_tr)

xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 400),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 400))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(6,5))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k')
plt.title("SVM with Polynomial Kernel (degree=3)")
plt.show()
```

7. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually

```python
import numpy as np, matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

X, y = make_classification(n_samples=500, n_features=2, n_redundant=0, n_informative=2,
                           class_sep=1.2, random_state=42)
Cs = [0.1, 1, 10, 100]
xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 400),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 400))

plt.figure(figsize=(10,10))
for i, C in enumerate(Cs, 1):
    clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", C=C, gamma="scale"))
    clf.fit(X, y)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax = plt.subplot(2,2,i)
    ax.contourf(xx, yy, Z, alpha=0.3)
    ax.scatter(X[:,0], X[:,1], c=y, s=10, edgecolor='k')
    ax.set_title(f"SVM RBF, C={C}")
plt.tight_layout(); plt.show()
```

11. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

param_grid = {"C":[0.1,1,10], "gamma":["scale", 0.1, 0.01], "kernel":["linear","rbf","poly"]}
grid = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1)
grid.fit(X_tr, y_tr)
print("Best params:", grid.best_params_)
print("Test accuracy:", accuracy_score(y_te, grid.best_estimator_.predict(X_te)))
```

16. Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results

17. Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy

18. Write a Python program to train an SVM Classifier using Linear, Polynomial, and RBF kernels on the Breast Cancer dataset and compare their accuracy

21. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC, LinearSVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

X, y = load_wine(return_X_y=True)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

base = Pipeline([("sc", StandardScaler()), ("svc", SVC(kernel="rbf", random_state=42))]).fit(X_tr, y_tr)
sel  = Pipeline([("sc", StandardScaler()),
                 ("rfe", RFE(LinearSVC(dual=False, max_iter=5000), n_features_to_select=8)),
                 ("svc", SVC(kernel="rbf", random_state=42))]).fit(X_tr, y_tr)

print("Without RFE acc:", accuracy_score(y_te, base.predict(X_te)))
print("With RFE acc   :", accuracy_score(y_te, sel.predict(X_te)))
```

```python
import seaborn as sns, matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

X, y = load_iris(return_X_y=True)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", random_state=42)).fit(X_tr, y_tr)
cm = confusion_matrix(y_te, clf.predict(X_te))

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cbar=False)
plt.title("SVM Confusion Matrix"); plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
```

25.  Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error

cal = fetch_california_housing()
X, y = cal.data, cal.target
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, random_state=42)

svr = make_pipeline(StandardScaler(), SVR(kernel="rbf", C=10.0))
svr.fit(X_tr, y_tr)
pred = svr.predict(X_te)
print("MAE:", mean_absolute_error(y_te, pred))
```

26. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score

X, y = load_breast_cancer(return_X_y=True)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

gnb = GaussianNB().fit(X_tr, y_tr)
proba = gnb.predict_proba(X_te)[:,1]
print("ROC-AUC:", roc_auc_score(y_te, proba))
```

27. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score

X, y = load_breast_cancer(return_X_y=True)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", probability=True, random_state=42)).fit(X_tr, y_tr)
scores = clf.predict_proba(X_te)[:,1]
prec, rec, thr = precision_recall_curve(y_te, scores)
ap = average_precision_score(y_te, scores)

plt.figure(figsize=(6,5))
plt.plot(rec, prec)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title(f"Precision-Recall Curve (AP={ap:.3f})")
plt.grid(True); plt.show()
```
